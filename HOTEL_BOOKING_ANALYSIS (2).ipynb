````python
# create a function which will give a bar chart of data  respective with a col.

def get_count_from_column_bar(df, column_label):
  df_hbpd = df[column_label].value_counts()
  df_hbpd = pd.DataFrame({'index':df_hbpd.index, 'count':df_hbpd.values})
  return df_hbpd


def plot_bar_chart_from_column(df, column_label, t1):
  df_hbpd = get_count_from_column_bar(df, column_label)
  fig, ax = plt.subplots(figsize=(14, 6))
  c= ['g','r','b','c','y']
  ax.bar(df_hbpd['index'], df_hbpd['count'], width = 0.4, align = 'edge', edgecolor = 'black', linewidth = 4, color = c, linestyle = ':', alpha = 0.5)
  plt.title(t1, bbox={'facecolor':'0.8', 'pad':3})
  plt.legend()
  plt.ylabel('Count')
  plt.xticks(rotation = 15) # use to format the lable of x-axis
  plt.xlabel(column_label)
  plt.show()


````

````python
# Chart - 1 visualization code

def get_count_from_column(df, column_label):
  df_hbpd = df[column_label].value_counts()
  df_hbpd = pd.DataFrame({'index':df_hbpd.index, 'count':df_hbpd.values})
  return df_hbpd

# plot a pie chart from grouped data
def plot_pie_chart_from_column(df, column_label, t1, exp):
  df_hbpd = get_count_from_column(df, column_label)
  fig, ax = plt.subplots(figsize=(14,9))
  ax.pie(df_hbpd.loc[:, 'count'], labels=df_hbpd.loc[:, 'index'], autopct='%1.2f%%',startangle=90,shadow=True, labeldistance = 1, explode = exp)
  plt.title(t1, bbox={'facecolor':'0.7', 'pad':6})
  ax.axis('equal')
  plt.legend()
  plt.show()
````

````python
exp1 = [0.05,0.05]
plot_pie_chart_from_column(booking_hotel_df, 'hotel', 'Booking percentage of Hotel by Name', exp1)
````
